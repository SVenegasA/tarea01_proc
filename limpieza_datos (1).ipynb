```python
# Ahora por cada df de participacion extraigo las keywords de su texto principal
from keybert import KeyBERT

kw_model = KeyBERT()

def get_keywords(text, top_n=3):
    # Devuelve lista de (keyword, score)
    keywords = kw_model.extract_keywords(text, top_n=top_n)
    return [k[0] for k in keywords]

df_participacion_2024["keywords"] = df_participacion_2024["TEXTO_PRINCIPAL"].apply(lambda x: get_keywords(x, top_n=8))
df_participacion_2024[["TEXTO_PRINCIPAL", "keywords"]].head()
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer(model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2"))  # Multilingüe (incluye español)

def transform_words(keywords):   
    
    # Convertir keywords a embeddings
    embeddings = model.encode(keywords, convert_to_tensor=True)
    
    # Calcular matriz de similitud
    cos_sim = util.pytorch_cos_sim(embeddings, embeddings)
    
    # Umbral para considerar duplicados (ajustable)
    threshold = 0.85
    
    # Filtrar duplicados semánticos
    final_keywords = []
    seen = set()
    
    for i in range(len(keywords)):
        if i in seen:
            continue
        final_keywords.append(keywords[i])
        for j in range(i + 1, len(keywords)):
            if cos_sim[i][j] > threshold:
                seen.add(j)
    
    return keywords

df_participacion_2023_1["keywords"] = df_participacion_2023_1["keywords"].apply(lambda x: transform_words(x))

```
