```python
def limpiar_texto(txt):
    txt = re.sub(r"\\n", " ", txt)
    txt = txt.lower()
    return txt.strip()
```

```python
# Ahora por cada df de participacion extraigo las keywords de su texto principal
from keybert import KeyBERT
from transformers import AutoModel, AutoTokenizer
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("PlanTL-GOB-ES/roberta-base-bne")
kw_model = KeyBERT(model)

def get_keywords(text, top_n=8):
    # Devuelve lista de (keyword, score)
    keywords = kw_model.extract_keywords(text, keyphrase_ngram_range = (1,30), stop_words = "spanish", use_maxsum=True, nr_candidates = 20, top_n=top_n)
    return [k[0] for k in keywords]
```

```python

df_participacion_2024["keywords"] = df_participacion_2024["TEXTO_PRINCIPAL"].apply(lambda x: get_keywords(x, top_n=8))
df_participacion_2024[["TEXTO_PRINCIPAL", "keywords"]].head()
df_participacion_2024.to_csv("participacion_2024.csv", index = False)
```

```python
from sentence_transformers import SentenceTransformer, util
import ast


model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")  # Multilingüe (incluye español)

df_participacion_2024["keywords"] = df_participacion_2024["keywords"].apply(ast.literal_eval)


def clean_keywords(keywords_list, threshold=0.85):
    """
    Elimina palabras clave semánticamente similares de una lista.
    """
    if not keywords_list or len(keywords_list) <= 1:
        return keywords_list
    
    embeddings = model.encode(keywords_list, convert_to_tensor=True)

    # Si solo hay un embedding, devolvemos directamente la lista
    if embeddings.shape[0] <= 1:
        return keywords_list

    cos_sim = util.pytorch_cos_sim(embeddings, embeddings)

    final_keywords = []
    seen = set()

    for i in range(len(keywords_list)):
        if i in seen:
            continue
        final_keywords.append(keywords_list[i])
        for j in range(i + 1, len(keywords_list)):
            if cos_sim[i][j].item() > threshold:
                seen.add(j)

    return final_keywords


df_participacion_2024["filtered_keywords"] = df_participacion_2024["keywords"].apply(clean_keywords)
```

```python
df_participacion_2023_1["keywords"] = df_participacion_2023_1["keywords"].apply(ast.literal_eval)
df_participacion_2023_2["keywords"] = df_participacion_2023_2["keywords"].apply(ast.literal_eval)

df_participacion_2023_1["filtered_keywords"] = df_participacion_2023_1["keywords"].apply(clean_keywords)
df_participacion_2023_2["filtered_keywords"] = df_participacion_2023_2["keywords"].apply(clean_keywords)
```
